## CLASSIFICATION PROBLEM 

```python
# make a binary class dataset 
from sklearn.datasets import make_classification
```

```python
x , y = make_classification(n_samples=1000,
                    n_features=3 ,
                    n_classes=2,
                    n_redundant=1,
                    random_state=999)
```

```python
x #indoendet feature 
```

```python
y # deoendednt feature -> output 
```

```python
# train test split 
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test  = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
# k neighbour classifier 
```

```python
from sklearn.neighbors import KNeighborsClassifier

# model create 
classifier = KNeighborsClassifier(n_neighbors=5,weights='distance',algorithm='auto',p=2)

# traine the model 
classifier.fit(x_train,y_train)

# predict 
y_test_pred = classifier.predict(x_test)
```

```python
# acuuracy check 
from sklearn.metrics import accuracy_score , confusion_matrix,classification_report

print(accuracy_score(y_test,y_test_pred))

print(confusion_matrix(y_test,y_test_pred))

print(classification_report(y_test,y_test_pred))
```

```python
# hyperparqametre tuning 

parameter = {
             'n_neighbors' :[3,4,5,6,7,8,9,10],
             'weights' :['uniform', 'distance'],
             'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],
             'p':[1,2]
            }
```

```python
from sklearn.model_selection import GridSearchCV

# model creation 
clf = GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',verbose=3,cv=3)

clf.fit(x_train,y_train)
```

```python
# best parameter 
clf.best_params_
```

```python
y_test_pred= clf.predict(x_test)
```
